## **1.**

### **Requisitos Funcionales (RF)**

1. **Gestión de la Página Web (RF01-RF04)**:
   - RF01: El sistema debe ofrecer una página web para promocionar el negocio.
   - RF02: La página web debe ser actualizada por un administrador del sitio.
   - RF03: El sistema debe permitir a una persona registrarse como cliente en unos pocos pasos fáciles desde la propia web.
   - RF04: Los clientes deben tener acceso 24x7 a secciones específicas orientadas a generar solicitudes de estampado.

2. **Solicitudes de Estampado (RF05-RF07)**:
   - RF05: Los clientes deben poder solicitar la impresión de logotipos en diferentes tipos de productos según disponibilidad en el almacén.
   - RF06: El sistema debe clasificar automáticamente a los clientes según el volumen solicitado en tres categorías: PERSONA, PYME, y EMPRESAPLUS, aplicando descuentos del 0%, 5%, y 10% respectivamente.
   - RF07: Los clientes EMPRESAPLUS pueden solicitar la participación del diseñador interno para optimizar su logotipo.

3. **Gestión de Pagos y Revisión (RF08-RF10)**:
   - RF08: Los clientes deben pagar antes de la ejecución de la orden, validando su tarjeta de crédito con un cargo simbólico.
   - RF09: Pedidos superiores a 10 prendas deben ser revisados y aprobados por un responsable en menos de 24 horas.
   - RF10: Los clientes pueden cancelar una orden antes de su ejecución; administradores y responsables también pueden cancelar órdenes.

4. **Gestión de Clientes y Pedidos (RF11-RF13)**:
   - RF11: Los clientes pueden actualizar su información, mientras que administradores necesitan permiso de responsables para modificar o bloquear cuentas.
   - RF12: Los clientes tienen acceso a su historial de pedidos y facturas, con opciones de búsqueda por fecha o producto y solicitud de facturas en papel.
   - RF13: Administrativos tienen acceso al listado de clientes y pedidos, y pueden imprimir facturas para enviarlas por correo.

5. **Gestión Interna (RF14-RF15)**:
   - RF14: Toda la información de usuarios, pedidos y facturas debe almacenarse por al menos 5 años y respaldarse dos veces por semana.
   - RF15: Los lunes deben generarse informes sobre el inventario y pedidos pendientes para planificar compras de material, que serán aprobadas por administrativos.

### **Requisitos No Funcionales (RNF)**

#### **RNF del Producto**
1. **Categoría de usabilidad**:
   - RNF01: El sistema debe ser intuitivo para clientes y administradores.
   - RNF02: La página web debe ser accesible para personas con discapacidad (accesibilidad).
   - RNF03: La interfaz debe estar disponible en español.

2. **Eficiencia**:
   - RNF04: El sistema debe permitir realizar solicitudes de estampado en menos de 5 segundos.
   - RNF05: Debe soportar un volumen de al menos 5000 solicitudes simultáneas (escalabilidad).

3. **Dependibilidad**:
   - RNF06: Disponibilidad del sistema garantizada al 99.9% mensual.
   - RNF07: Copias de seguridad de los datos realizadas dos veces por semana.
   - RNF08: Garantizar que todos los pedidos y datos de clientes se mantienen consistentes (integridad).

4. **Seguridad**:
   - RNF09: Restricción de acceso según roles (administradores, responsables, clientes).
   - RNF10: Cumplimiento con normativas como GDPR y otras leyes de protección de datos.
   - RNF11: Todos los accesos y modificaciones deben ser auditables.

#### **RNF de la Organización**
1. **Entorno**:
   - RNF12: El sistema debe operar en infraestructura existente en la imprenta.

2. **Organizacionales**:
   - RNF13: Administradores y responsables deben poder acceder al sistema desde dispositivos móviles.

3. **Desarrollo**:
   - RNF14: Uso de estándares de desarrollo ágil y pruebas automatizadas.

#### **RNF Externos**
1. **Regulatorios**:
   - RNF15: Cumplir con las normativas legales locales para almacenamiento de datos (5 años mínimo).

2. **Legislativos**:
   - RNF16: Las facturas deben cumplir con los estándares contables nacionales.

## **2.**

### **Posibles Conflictos**

1. **RNF04 vs RF08**:
   - RNF04 establece que las solicitudes deben ser rápidas (< 5 segundos), pero RF08 requiere validación de pago, lo que puede depender de terceros y generar retrasos.

2. **RF10 vs RNF09**:
   - Permitir a varios roles cancelar órdenes puede generar conflictos si no se establece una jerarquía clara en los permisos de cancelación.

## **4.**

### **Priorización de RF**
#### Justificación de las 6 prioridades más importantes:

1. **RF05 (Solicitudes de estampado)**:
   - Es el núcleo del negocio. Permite que clientes realicen pedidos directamente en línea.

2. **RF06 (Clasificación de clientes)**:
   - Optimiza la relación con clientes según el volumen, incentivando descuentos para mayores volúmenes.

3. **RF08 (Pagos validados previamente)**:
   - Evita el riesgo financiero al garantizar que los pedidos sean prepagados.

4. **RF09 (Revisión de pedidos grandes)**:
   - Asegura control de calidad y evita errores en pedidos grandes, clave para mantener la reputación del negocio.

5. **RF12 (Historial de pedidos y facturas)**:
   - Mejora la transparencia y confianza con los clientes al permitir acceso a su información.

6. **RF14 (Almacenamiento y respaldos)**:
   - Cumple con requisitos legales y garantiza continuidad del negocio en caso de desastres.
